```python
X_train = df_train.drop('price', axis=1)
y_train = df_train['price']
X_test = df_test.copy()
```

```python
X_train = pd.get_dummies(X_train, drop_first=True, dtype=int)
X_test = pd.get_dummies(X_test, drop_first=True, dtype=int)
```

```python
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = DecisionTreeRegressor()
```

```python
param_grid = {
   'criterion': ['squared_error', 'absolute_error'],
   'max_depth': [5, 10, 15, 20, None],
   'min_samples_split': [2, 10, 20],
   'min_samples_leaf': [1, 5, 15]
}
```

```python
final_model = GridSearchCV(model, param_grid, cv=5, verbose = 3)
```

```python
y_train
```

```python
final_model.fit(X_train, y_train)
```

```python
y_pred = final_model.predict(X_test)
```

```python
df_sub['price'] = y_pred
```

```python
df_sub.to_csv('sub.csv', index=False, index_label=False)
```

**RANDOM FOREST REGRESSOR**

```python
X_train = df_train.drop('price', axis=1)
y_train = df_train['price']
X_test = df_test.copy()
```

```python
X_train = pd.get_dummies(X_train, drop_first=True, dtype=int)
X_test = pd.get_dummies(X_test, drop_first=True, dtype=int)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = RandomForestRegressor()
```

```python
params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5],
    'max_features': [1.0, 'sqrt']
}
```

```python
rf_final_model = GridSearchCV(model, params, cv=5, verbose=3)
```

```python
rf_final_model.fit(X_train, y_train)
```

```python
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)
```

```python
rf_y_pred = rf_final_model.predict(X_test)
```

```python
df_sub['price'] = rf_y_pred
```

```python
df_sub.to_csv('rf_sub.csv', index=False, index_label=False)
```

**Score: 1891.7341**
